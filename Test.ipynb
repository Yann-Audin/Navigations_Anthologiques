```python
# Scrapping the Gutenberg Project
data = pd.read_csv("query_2022_02_25.csv")
texts = []
ids = []
startBooks = ["*** START OF THIS PROJECT GUTENBERG EBOOK", "START OF THE PROJECT GUTENBERG EBOOK", "START OF THIS PROJECT GUTENBERG EBOOK"]
endBooks = ["*** END OF THIS PROJECT GUTENBERG EBOOK", "*** END OF THE PROJECT GUTENBERG EBOOK", "End of the Project Gutenberg EBook", 
            "***END OF THE PROJECT GUTENBERG EBOOK", "***END OF THE PROJECT GUTENBERG EBOOK"]
for id in data.gutenbergid:
    time.sleep(0.5)
    path = "https://www.gutenberg.org/files/" + str(id) + "/" + str(id) + "-0.txt"
    #https://www.gutenberg.org/files/1400/1400-0.txt
    #print(path)
    page = requests.get(path)
    text = page.text
    if text.find("<!DOCTYPE html>") == -1:
        start = -1
        for startBook in startBooks:
            tempStart = text.find(startBook)
            if tempStart != -1:
                start = tempStart + 5
        if start == 4:
            print(id)
        end = -1
        for endBook in endBooks:
            tempEnd = text.find(endBook)
            if tempEnd != -1:
                end = tempEnd
        if end == -1:
            print(id)
        textLit = text[start:end]
        startUp = textLit.find("***") + 3
        textUp = textLit[startUp:len(textLit)]
        texts.append(textUp)
        ids.append(id)
df = pd.DataFrame(data=texts, index=ids, columns=["texts"])
df.to_csv("text.csv")
df = pd.read_csv("texts.csv")
```
